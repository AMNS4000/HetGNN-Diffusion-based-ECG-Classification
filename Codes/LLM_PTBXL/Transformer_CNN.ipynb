```python
import numpy as np
import pywt
from scipy.fft import fft

def extract_wavelet_features(ecg_signal, wavelet='db4', level=4):
    """Extract wavelet coefficients for each lead."""
    coeffs = pywt.wavedec(ecg_signal, wavelet, level=level)
    return np.concatenate([coeff for coeff in coeffs])

def extract_fourier_features(ecg_signal):
    """Extract Fourier Transform features (magnitude of FFT) for each lead."""
    fourier_transformed = np.abs(fft(ecg_signal))
    return fourier_transformed

def concatenate_features_with_ecg(ecg_data):
    """Concatenate original ECG data with wavelet and Fourier features."""
    num_samples, num_leads, seq_len, _ = ecg_data.shape
    
    # Placeholder for the final concatenated data
    extended_data = []
    
    for sample in range(num_samples):
        leads_data = []
        for lead in range(num_leads):
            ecg_signal = ecg_data[sample, lead, :, 0]
            
            # Extract features
            wavelet_features = extract_wavelet_features(ecg_signal)
            fourier_features = extract_fourier_features(ecg_signal)
            
            # Concatenate the original ECG signal, wavelet, and Fourier features
            concatenated_features = np.concatenate((ecg_signal, wavelet_features, fourier_features), axis=0)
            
            # Add to the leads data
            leads_data.append(concatenated_features.reshape(1, -1))
        
        # Stack all leads together
        leads_data = np.stack(leads_data, axis=0)
        extended_data.append(leads_data)
    
    # Convert list to numpy array
    extended_data = np.stack(extended_data, axis=0)
    
    # Final shape will be (num_samples, num_leads, 1000 + wavelet + fourier, 1)
    return extended_data.reshape(num_samples, num_leads, -1, 1)

# Assuming x_train is your original training data of shape (19234, 12, 1000, 1)
x_train_extended = concatenate_features_with_ecg(x_train)

print("New shape after adding wavelet and Fourier features:", x_train_extended.shape)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
import numpy as np

# Dataset definition
class ECGDataset(Dataset):
    def __init__(self, data, labels):
        self.data = torch.tensor(data, dtype=torch.float32)  # Convert data to float32
        self.labels = torch.tensor(labels, dtype=torch.float32)  # Convert labels to float32

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx].transpose(0, 1), self.labels[idx]

# Hybrid CNN-Transformer Model
class CNNTransformerHybrid(nn.Module):
    def __init__(self, num_classes, num_leads=12, seq_len=3067, d_model=128, num_heads=4, num_transformer_layers=2):
        super(CNNTransformerHybrid, self).__init__()

        # CNN layers for ECG leads
        self.conv1 = nn.Conv1d(in_channels=num_leads, out_channels=32, kernel_size=7, padding=3)
        self.conv2 = nn.Conv1d(in_channels=32, out_channels=64, kernel_size=5, padding=2)
        self.conv3 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.pool = nn.MaxPool1d(kernel_size=2, stride=2)

        # Transformer Encoder
        cnn_output_size = seq_len // 8  # After 3 CNN layers with pooling, the sequence length is reduced
        self.pos_encoder = nn.Parameter(torch.zeros(1, cnn_output_size, d_model))
        encoder_layers = nn.TransformerEncoderLayer(d_model=d_model, nhead=num_heads, dim_feedforward=256, dropout=0.1)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layers, num_layers=num_transformer_layers)

        # Fully connected layers
        self.fc1 = nn.Linear(128, 256)  # Input matches the output of the last CNN layer
        self.fc2 = nn.Linear(256, 256)
        self.fc3 = nn.Linear(256, num_classes)

    def forward(self, x):
        # Reshape the input (batch_size, 12, 3067, 1) to (batch_size, 12, 3067)
        ecg_data = x.squeeze(-1)  # Shape: (batch_size, 12, 3067)
        
        # Change shape to (batch_size, channels, sequence_length) for CNN layers
        x = ecg_data.permute(0, 2, 1)  # Shape: (batch_size, 12, 3067)
        
        # Now pass through the CNN layers
        x = self.conv1(x)  # (batch_size, 32, 3067)
        x = F.relu(x)
        x = self.pool(x)  # (batch_size, 32, 1533)
        
        x = self.conv2(x)  # (batch_size, 64, 1533)
        x = F.relu(x)
        x = self.pool(x)  # (batch_size, 64, 766)
        
        x = self.conv3(x)  # (batch_size, 128, 766)
        x = F.relu(x)
        x = self.pool(x)  # (batch_size, 128, 383)

        # Reshape for Transformer: (batch_size, seq_len//8, 128)
        x = x.permute(0, 2, 1)  # Shape: (batch_size, 383, 128)

        # Add positional encoding
        x = x + self.pos_encoder[:, :x.size(1)]

        # Transformer Encoder
        x = self.transformer_encoder(x)
        x = torch.mean(x, dim=1)  # Aggregate sequence length dimension

        # Fully connected layers
        x = self.fc1(x)  # (batch_size, 256)
        x = F.relu(x)
        x = F.dropout(x, p=0.5)
        x = self.fc2(x)  # (batch_size, 256)
        x = F.relu(x)
        x = F.dropout(x, p=0.5)
        x = self.fc3(x)  # (batch_size, num_classes)

        return x



# Initialize weights to avoid instability
def init_weights(m):
    if isinstance(m, nn.Conv1d) or isinstance(m, nn.Linear):
        nn.init.xavier_uniform_(m.weight)
        if m.bias is not None:
            nn.init.constant_(m.bias, 0)

# Data preparation
def prepare_data(data, raw_labels, validation_split=0.2, batch_size=32):
    # Split the dataset into training and validation sets
    train_data, val_data, train_labels, val_labels = train_test_split(data, raw_labels, test_size=validation_split, random_state=42)

    # Create datasets
    train_dataset = ECGDataset(train_data, train_labels)
    val_dataset = ECGDataset(val_data, val_labels)

    # Create dataloaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, val_loader

# Model training function
def train_model(model, train_loader, val_loader, num_epochs=20, learning_rate=0.001):
    criterion = nn.BCEWithLogitsLoss()  # Suitable for multi-label classification
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for inputs, labels in train_loader:
            if torch.cuda.is_available():
                inputs, labels = inputs.cuda(), labels.cuda()

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels.float())
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * inputs.size(0)

        epoch_loss = running_loss / len(train_loader.dataset)
        print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss:.4f}')

        # Validation
        model.eval()
        val_loss = 0.0
        all_preds = []
        all_labels = []
        with torch.no_grad():
            for inputs, labels in val_loader:
                if torch.cuda.is_available():
                    inputs, labels = inputs.cuda(), labels.cuda()

                outputs = model(inputs)
                loss = criterion(outputs, labels.float())
                val_loss += loss.item() * inputs.size(0)

                all_preds.append(outputs.sigmoid().cpu())
                all_labels.append(labels.cpu())

        val_loss = val_loss / len(val_loader.dataset)
        print(f'Validation Loss: {val_loss:.4f}')

        # F1 Score for multi-label classification
        all_preds = torch.cat(all_preds)
        all_labels = torch.cat(all_labels)
        preds = (all_preds > 0.5).float()
        f1 = f1_score(all_labels, preds, average='macro')
        print(f'Validation F1 Score: {f1:.4f}')


if __name__ == "__main__":
    data = x_train_extended  # Assuming ECG data is loaded (shape: (19234, 12, 3067, 1))
    labels = y_train  # Corresponding labels (shape: (19234, num_classes))

    train_loader, val_loader = prepare_data(data, labels, validation_split=0.2, batch_size=64)
    model = CNNTransformerHybrid(num_classes=5)  # Assuming 5 classes for multi-label classification

    # Apply weight initialization
    model.apply(init_weights)

    if torch.cuda.is_available():
        model = model.cuda()

    train_model(model, train_loader, val_loader, num_epochs=20)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
import numpy as np

possible_labels = ['STTC', 'MI', 'CD', 'HYP', 'NORM', '']
num_classes = len(possible_labels)

label_to_idx = {label: i for i, label in enumerate(possible_labels)}

def encode_labels(label_list, num_classes):
    encoding = np.zeros(num_classes)
    for label in label_list:
        if label in label_to_idx:
            encoding[label_to_idx[label]] = 1
    return encoding

class ECGDataset(Dataset):
    def __init__(self, data, labels):
        self.data = torch.tensor(data, dtype=torch.float32)  # Convert data to float32
        self.labels = torch.tensor(labels, dtype=torch.float32)  # Convert labels to float32

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        # Remove the extra feature dimension using squeeze
        return self.data[idx].squeeze(-1).transpose(0, 1), self.labels[idx]



class CNNTransformerHybrid(nn.Module):
    def __init__(self, num_classes, num_leads=12, seq_len=1000, d_model=128, num_heads=4, num_transformer_layers=2, num_neurokit_features=15):
        super(CNNTransformerHybrid, self).__init__()

        # CNN layers
        self.conv1 = nn.Conv1d(in_channels=num_leads, out_channels=32, kernel_size=7, padding=3)
        self.conv2 = nn.Conv1d(in_channels=32, out_channels=64, kernel_size=5, padding=2)
        self.conv3 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.pool = nn.MaxPool1d(kernel_size=2, stride=2)
        
        # Transformer Encoder
        self.pos_encoder = nn.Parameter(torch.zeros(1, seq_len//8, d_model))
        encoder_layers = nn.TransformerEncoderLayer(d_model=d_model, nhead=num_heads, dim_feedforward=256, dropout=0.1)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layers, num_layers=num_transformer_layers)
        
        # Fully connected layers
        self.fc1 = nn.Linear(d_model + num_neurokit_features, 128)
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        # Input shape: (batch_size, num_leads, seq_len)

        # Separate ECG data and NeuroKit features
        ecg_data = x
        ecg_data = ecg_data.permute(0,2,1)

        # CNN layers
        x = self.conv1(ecg_data)  # (batch_size, 32, seq_len)
        x = nn.ReLU()(x)
        x = self.pool(x)   # (batch_size, 32, seq_len//2)

        x = self.conv2(x)  # (batch_size, 64, seq_len//2)
        x = nn.ReLU()(x)
        x = self.pool(x)   # (batch_size, 64, seq_len//4)

        x = self.conv3(x)  # (batch_size, 128, seq_len//4)
        x = nn.ReLU()(x)
        x = self.pool(x)   # (batch_size, 128, seq_len//8)

        # Reshape for Transformer: (batch_size, seq_len//8, d_model)
        x = x.permute(0, 2, 1)  # (batch_size, seq_len//8, 128)

        # Positional Encoding
        seq_len_transformed = x.size(1)
        pos_enc = self.pos_encoder[:, :seq_len_transformed, :]
        x = x + pos_enc  # Add positional encoding

        # Transformer Encoder
        x = self.transformer_encoder(x)  # (batch_size, seq_len//8, d_model)

        # Global Average Pooling
        x = torch.mean(x, dim=1)  # (batch_size, d_model)

        x = x.view(x.size(0), -1)  # Flatten to (batch_size, 128 * (seq_len//8))
        # Fully connected layers
        x = self.fc1(x)  # (batch_size, 128)
        x = nn.ReLU()(x)
        x = nn.Dropout(p=0.5)(x)
        x = self.fc2(x)  # (batch_size, num_classes)

        return x





def prepare_data(data, raw_labels, validation_split=0.2, batch_size=32, sampling_rate=1000):
    # Split the dataset into training and validation sets
    train_data, val_data, train_labels, val_labels = train_test_split(data, raw_labels, test_size=validation_split, random_state=42)

    # Create datasets
    train_dataset = ECGDataset(train_data, train_labels)
    val_dataset = ECGDataset(val_data, val_labels)

    # Create dataloaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, val_loader



# Training function
def train_model(model, train_loader, val_loader, num_epochs=20, learning_rate=0.001):
    # Loss and optimizer
    criterion = nn.BCEWithLogitsLoss()  # Suitable for multi-label classification
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    # Training loop
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for inputs, labels in train_loader:
            # Move data to GPU if available
            if torch.cuda.is_available():
                inputs, labels = inputs.cuda(), labels.cuda()

            # Zero the parameter gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, labels.float())

            # Backward pass and optimize
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)

        epoch_loss = running_loss / len(train_loader.dataset)
        print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss:.4f}')

        # Validation phase
        model.eval()
        val_loss = 0.0
        all_preds = []
        all_labels = []
        with torch.no_grad():
            for inputs, labels in val_loader:
                if torch.cuda.is_available():
                    inputs, labels = inputs.cuda(), labels.cuda()

                outputs = model(inputs)
                loss = criterion(outputs, labels.float())
                val_loss += loss.item() * inputs.size(0)

                all_preds.append(outputs.sigmoid().cpu())
                all_labels.append(labels.cpu())

        val_loss = val_loss / len(val_loader.dataset)
        print(f'Validation Loss: {val_loss:.4f}')

        # Calculate F1 Score for multi-label classification
        all_preds = torch.cat(all_preds)
        all_labels = torch.cat(all_labels)
        preds = (all_preds > 0.5).float()
        f1 = f1_score(all_labels, preds, average='macro')
        print(f'Validation F1 Score: {f1:.4f}')

# Example usage
if __name__ == "__main__":
    # Prepare data
    train_loader, val_loader = prepare_data(x_train,y_train,validation_split=0.2, batch_size=32)

    # Initialize model
    model = CNNTransformerHybrid(num_classes=num_classes)
    
    # Move model to GPU if available
    if torch.cuda.is_available():
        model = model.cuda()

    # Train the model
    train_model(model, train_loader, val_loader, num_epochs=20)
```
